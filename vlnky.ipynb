```python
import matplotlib.pyplot as plt
import numpy as np


plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

x = np.arange(0, 20.0, 0.1)
```

```python
y_m = 2
omega = 2
fi_0 = 0.25* 3.14


vychylka = True
rychlost = True
zrychleni = True
energie_kineticka = False
energie_potencialni = False

```

```python
fig = plt.figure()
fig.set_figwidth(16)
fig.set_figheight(4)
ax = fig.add_subplot(111)

vychylka_plot = y_m * np.sin(omega*x + fi_0)
rychlost_plot = omega * y_m * np.cos(omega*x + fi_0)
zrychleni_plot = - omega**2 * y_m * np.sin(omega*x + fi_0)
y = np.zeros(len(x))
ax.plot(x,y)

pruznost = 1    

if vychylka: ax.plot(x, vychylka_plot, label='okamžitá výchylka')
if rychlost: ax.plot(x, rychlost_plot, label = 'okamžitá rychlost')
if zrychleni: ax.plot(x, zrychleni_plot, label = 'okamžité zrychlení')
ax.legend()
unit = 0.5

x_tick = np.arange(0, 6 + unit, unit)
x_label = [ r"$0$", r"$\frac{\pi}{2}$", r"$\pi$", r"$\frac{3\pi}{2}$", r"$2\pi$", r"$\frac{5\pi}{2}$", r"$3\pi$", r"$\frac{7\pi}{2}$", r"$4\pi$", r"$\frac{9\pi}{2}$", r"$5\pi$", r"$\frac{11\pi}{2}$", r"$6\pi$"]

ax.set_xticks(x_tick * np.pi)
ax.set_xticklabels(x_label, fontsize=10)

plt.grid()
plt.show()
```
